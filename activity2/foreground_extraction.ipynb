```python
from tqdm import tqdm
import os
import cv2
import numpy as np

#path of the folder containing videos of a particular activity
train_data='D:\\wave2'
for i in tqdm(os.listdir(train_data)):
       
        path=os.path.join(train_data,i)
        cam = cv2.VideoCapture(path)
        
        #creating a new folder to store every frame
        try: 
    
            loc='D:\\wave2mi'
            if not os.path.exists(loc): 
                os.makedirs(loc) 
  
        except OSError: 
            print ('Error: Creating directory of data') 
  
        currentframe = 0
  
        while(True): 
      
            ret,frame = cam.read() 
 
            if ret: 
       
                name = loc+"\\"+i+"mask" + str(currentframe) +'+wave2'+ '.jpg'
                print ('Creating...' + name) 
       
                cv2.imwrite(name, frame) 

                currentframe += 1
            else: 
                break
  

        cam.release() 
        cv2.destroyAllWindows() 


```

```python
#path of the folder which contains frames with exactly same background
for i in tqdm(os.listdir("D:\\www")):
    
        path=os.path.join("D:\\www",i)
        img=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
        img=cv2.resize(img,(64,64))
        
        #reading the image of the background alone
        image2 = cv2.imread("D:\\xxx\\bg_046.avimask0+box.jpg",cv2.IMREAD_GRAYSCALE)
        image2=cv2.resize(image2,(64,64))
        
        #substracting the background from the actual image to get just the human body
        img=cv2.subtract(image2,img)
        
        #image thresholding to get binary image
        blur = cv2.GaussianBlur(img,(1,1),0)
        ret3,th3 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        
        #saving all the images in a folder
        f="D:\\abd\\"+i
        cv2.imwrite(f+'.jpg',th3)


```
