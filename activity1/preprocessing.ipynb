```python
from tqdm import tqdm
import os
import cv2
import numpy as np
from moviepy.editor import VideoFileClip

#address of the folder containg videos of a particular activity
train_data='D:\\walk'

#storing the duration of video
for i in tqdm(os.listdir(train_data)):
        path=os.path.join(train_data,i)
       
        clip = VideoFileClip(path)
        c=clip.duration*2
        
        # Read the video from specified path 
        cam = cv2.VideoCapture(path) 
        
        #defining background subtraction function to extract the moving part
        fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()
        
        try: 
            a=i.strip('.')
    
            loc='D:\\walkmi\\'+a
            if not os.path.exists(loc): 
                os.makedirs(loc) 
                
        # if not created then raise error 
        except OSError: 
            print ('Error: Creating directory of data') 
        # frame 
        currentframe = 0
  
        while(True): 
      
        # reading from frame 
            ret,frame = cam.read() 
 
            if ret: 
                mask = fgbg.apply(frame)
        
                #dialation to join certain broken parts of image
                mask = cv2.dilate(mask,None,iterations=2)
        
                name = loc+"\\mask" + str(currentframe) + '.jpg'
                print ('Creating...' + name) 
                
                #intensity function to give more weight to more recent frames
                mask=mask/(c*c*c)
                c=c-0.04
            
                # writing the extracted images 
                cv2.imwrite(name, mask) 
   
                currentframe += 1
            else: 
                break
  
        # Release all space and windows once done 
        cam.release() 
        cv2.destroyAllWindows() 


```

```python
from tqdm import tqdm
import os
import cv2
import numpy as np

#Address of folder containing different binary frames
train_data='D:\\walkmi'
for i in tqdm(os.listdir(train_data)):
        path=os.path.join(train_data,i)
        blended=0
        
        # creating a new folder to store the generated binary motion images
        try:     
            if not os.path.exists('D:\\walkbmi'): 
                os.makedirs('D:\\kthjogbmi') 
        except OSError: 
                print ('Error: Creating directory of data') 
                
        #adding all the images of different intensities for a particular video
        for j in tqdm(os.listdir(path)):
            path1=os.path.join(path,j)
            img=cv2.imread(path1,cv2.IMREAD_GRAYSCALE)
            img=cv2.resize(img,(64,64))
            blended+=img
            
        #storing the bmi in a different folder
        f="D:\\walkbmi\\"+i
        cv2.imwrite(f+'.jpg',blended)


```
